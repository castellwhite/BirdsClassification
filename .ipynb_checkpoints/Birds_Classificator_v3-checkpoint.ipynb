```python
# Resize images to 256X256
from PIL import Image

xsize = 224
ysize = 224

paths=['egret','mandarin','owl','puffin','toucan','wood_duck']

pre=''
birdKind=-1

xTrainL=[]
yTrainL=[]
xTestL=[]
yTestL=[]

for j in paths[:]:
    if(j=='egret'):
        pre='/egr'
        birdKind=0
    elif(j=='mandarin'):
        pre='/man'
        birdKind=1
    elif(j=='owl'):
        pre='/owl'
        birdKind=2
    elif(j=='puffin'):
        pre='/puf'
        birdKind=3
    elif(j=='wood_duck'):
        pre='/wod'
        birdKind=5
    elif(j=='toucan'):
        pre='/tou'
        birdKind=4
        
    counter=0
    for i in formatedList[:]:
        img = Image.open('birds/'+j+pre+i+'.jpg')
        wpercent = (basewidth/float(img.size[0]))
        img = img.resize((basewidth,hsize), Image.ANTIALIAS)
        img.save('birds/'+j+pre+i+'.jpg')
        
        img = load_img('birds/'+j+pre+i+'.jpg',grayscale=False,target_size=(xsize,ysize),interpolation='nearest') 
        x= img_to_array(img)
        
        if counter < 50:
            xTrainL.append(x)
            yTrainL.append(birdKind)
        
        else: 
            xTestL.append(x)
            yTestL.append(birdKind)
            
        counter=counter+1
        
```

```python
xTrain=np.asarray(xTrainL)
yTrainList=np.asarray(yTrainL)

xTest=np.asarray(xTestL)
yTestList=np.asarray(yTestL)

from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils

# encode class values as integers
encoder = LabelEncoder()
encoder.fit(yTrainList)
encoded_Y = encoder.transform(yTrainList)
# convert integers to dummy variables (i.e. one hot encoded)
yTrain = np_utils.to_categorical(encoded_Y)

# encode class values as integers
encoder = LabelEncoder()
encoder.fit(yTestList)
encoded_Y = encoder.transform(yTestList)
# convert integers to dummy variables (i.e. one hot encoded)
yTest = np_utils.to_categorical(encoded_Y)
```

```python
# Data path information.
trainDataPath= 'datasets/train'
testDataPath= 'datasets/test'

# Images dimensions.
widthSize =  224
heightSize = 224

# Batch Size.

batchSize = 16
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K

outputClasses = 6

if K.image_data_format() == 'channels_first':
    input_shape = (3, widthSize, heightSize)
else:
    input_shape = (widthSize, heightSize, 3)

model = Sequential()
NumN=32
NumN2=64

model.add(Conv2D(NumN, (3, 3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(NumN, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(NumN2, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(NumN2))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(outputClasses))
model.add(Activation('softmax'))

model.summary()
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K

outputClasses = 6

if K.image_data_format() == 'channels_first':
    input_shape = (3, widthSize, heightSize)
else:
    input_shape = (widthSize, heightSize, 3)

model = Sequential()
NumN=32
NumN2=64

model.add(Flatten(input_shape=input_shape))

model.add(Dense(outputClasses))
model.add(Activation('softmax'))

model.summary()
```

```python
# Compile the convolutional neural network
from keras.optimizers import adam
from keras.optimizers import SGD
from keras.optimizers import adamax
from keras.optimizers import RMSprop


op = RMSprop(lr=0.0001)
model.compile(loss='categorical_crossentropy',
             optimizer=op,
              metrics=['accuracy'])
```

```python
batch_size = 32
epochs = 3

history = model.fit(xTrain, yTrain,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data = [xTest,yTest])
```
